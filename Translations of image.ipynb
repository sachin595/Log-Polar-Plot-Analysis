```python
import cv2
import numpy as np

# Load the centered image
img = cv2.imread("C:/Users/DELL/Desktop/CVS/Log Polar/pics/car.jpg")


# Calculate the shift parameters
rows, cols = img.shape[:2]
shift_x = int(cols / 4.5)
shift_y = int(rows / -3.5)

# Define the transformation matrix
M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])

# Apply the transformation to the image
result = cv2.warpAffine(img, M, (cols, rows))

# Save the result
cv2.imwrite('non_centered_img.jpg', result)

from IPython import display
display.Image("non_centered_img.jpg")
```
